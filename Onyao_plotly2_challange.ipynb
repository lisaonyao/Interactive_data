```python
import pandas as pd
import plotly.express as px
```

```python
cricket = pd.read_csv('cricket.csv')
cricket.head()
```

```python
fig1 = px.scatter(cricket,\
                 x='Innings',\
                 y='Runs',\
                 height=500,\
                 title="Cricket players statistics")

fig1.show()
```

```python
fig2 = px.scatter(cricket,\
                 x='Innings',\
                 y='Runs',\
                 height=500,\
                 title="Cricket players statistics",
                 hover_name="Player",
                 hover_data=["Country"])

fig2.show()
```

```python
fig3 = px.scatter(cricket,\
                 x='Innings',\
                 y='Runs',\
                 height=500,\
                 title="Cricket players statistics",
                 hover_name="Player",
                 hover_data=["Country"])

button1 = {'method': 'update',
          'label':'Scatterplot'    ,
          'args': [
              {'type':'scatter'   },
              ]
          } 

button2 = {'method': 'update',
          'label':'Bar Graph'    ,
          'args': [
              { 'type':'bar'   },
              ]
          } 

fig3.update_layout(
    updatemenus=[{
        'type': 'buttons',
        'direction': "left",
        'pad': {"r": 10, "t": 10},
        'showactive': True,
        'x': 0.1,
        'xanchor': "left",
        'y': 1,
        'yanchor': "top",
        'buttons': [button1, button2]
    }]
)

fig3.show()
```

```python
fig4 = px.scatter(cricket,\
                 x='Innings',\
                 y='Runs',\
                 height=500,\
                 title="Cricket players statistics",
                 hover_name="Player",
                 hover_data=["Country"])
button3 = {'method': 'update',
          'label':  'Runs'  ,
          'args': [
              {  'y':[cricket['Runs']]   },
              {   'yaxis.title':'Runs' }]
          } 

button4 = {'method': 'update',
          'label':'No Outs'    ,
          'args': [
              {  'y':[cricket['No Outs']]  },
              {  'yaxis.title':'No Outs'  }]
          } 

fig4.update_layout(
    updatemenus=[{
        'type': 'buttons',
        'direction': "left",
        'pad': {"r": 10, "t": 10},
        'showactive': True,
        'x': 0.1,
        'xanchor': "left",
        'y': 1,
        'yanchor': "top",
        'buttons': [button3, button4]
    }]
)

fig4.show()
```

```python
fig5 = px.scatter(cricket,\
                 x='Innings',\
                 y='Runs',\
                 height=500,\
                 title="Cricket players statistics",
                 hover_name="Player",
                 hover_data=["Country"])

df_india= cricket[cricket['Country']=='INDIA']

button5 = {'method': 'update',
          'label':  'All Teams'  ,
          'args': [
              {  'x':[cricket['Innings']]   },
              {  'y':[cricket['Runs']] }]
          } 

button6 = {'method': 'update',
          'label':'India'    ,
          'args': [
              {  'x':[df_india['Innings']]  },
              {  'y':[df_india['Runs']]  }]
          } 

fig5.update_layout(
    updatemenus=[{
        'type': 'buttons',
        'direction': "left",
        'pad': {"r": 10, "t": 10},
        'showactive': True,
        'x': 0.1,
        'xanchor': "left",
        'y': 1,
        'yanchor': "top",
        'buttons': [button5, button6]
    }]
)

fig5.show()
```

```python
spotify=pd.read_csv('spotify.csv')
spotify.head()
```

```python
spotify= spotify.drop(columns=["Unnamed: 0"])
```

```python
spotify.head()
```

```python
fig = px.line(spotify,\
             x='endDate',\
             y='msPlayed',\
             title='Spotify Listening Time by user per day')
fig.show()
```

```python
fig = px.line(spotify,\
             x='endDate',\
             y='msPlayed',\
              color='userID',\
             title='Spotify Listening Time by user per day')
fig.show()
```

```python
fig= px.area(spotify,\
            x='endDate',\
            y='msPlayed',\
            color='userID',\
            title='Spotify Listening Time by user per day')
fig.show()
```

```python
fig= px.area(spotify,\
            x='day',\
            y='msPlayed',
            animation_frame='month',\
            animation_group='endDate',\
            color='userID',\
            title='Spotify Listening Time by user per day')
fig.show()
```

```python
fig= px.area(spotify,\
            x='day',\
            y='msPlayed',
            animation_frame='month',\
            animation_group='endDate',\
            color='userID',\
            title='Spotify Listening Time by user per day')
fig["layout"].pop("updatemenus")
fig.show()
```

```python
fig= px.area(spotify,\
            x='day',\
            y='msPlayed',
            animation_frame='month',\
            animation_group='endDate',\
             range_x= [1,31],\
             range_y=[0,55000000],\
            color='userID',\
            title='Spotify Listening Time by user per day')
fig["layout"].pop("updatemenus")
fig.show()
```

```python
fig= px.area(spotify,\
            x='day',\
            y='msPlayed',
            animation_frame='month',\
            animation_group='endDate',\
             range_x= [1,31],\
             range_y=[0,55000000],\
            color='userID',\
            title='Spotify Listening Time by user per day')
fig["layout"].pop("updatemenus")
fig.update_layout(sliders=[dict(bgcolor='green', bordercolor='black')])
fig.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from ipywidgets import widgets
import datetime
import seaborn as sns
from IPython.display import display
```

```python
%matplotlib inline
plt.rcParams['grid.linestyle'] = '--' # For better interactivity
```

```python
cgf = pd.read_csv('CardioGoodFitness.csv')
cgf.head()
```

```python
widgets.FloatSlider(
    value = cgf['Income'].min(),\
    min = cgf['Income'].min(), max = cgf['Income'].max(),\
    step = 1000, orientation = 'horizontal', description = 'Income Filter:',
    style = {'description_width': 'initial'}
)
```

```python
product_dropdown = widgets.Dropdown(
    options = cgf.columns.tolist(),\
    value = 'Gender', description = 'Attributes:',\
    disabled = False  
)
display(product_dropdown)
```

```python
y_variables = [col for col in cgf.columns if col not in ['Product', 'Income']]

widgets.Dropdown(
    options = y_variables,
    value = 'Age',
    description = ' y variable:'
)
```

```python
y_variable = 'Age'
plt.figure(figsize=(10, 6))
sns.countplot(x='Product', hue=y_variable, data=cgf, palette='viridis')
plt.title('Count of Products Grouped by Age')
plt.xlabel('Product')
plt.ylabel('Count')
plt.legend(title=y_variable)
plt.show()
```

```python
def update_chart(y_variable):
    plt.figure(figsize = (12,8))
    sns.countplot(x = 'Product', hue = y_variable,\
              data = cgf, palette = 'viridis')
    plt.title('Count of Products Grouped by{}'.format(y_variable))
    plt.xlabel('Product')
    plt.ylabel('Count')
    plt.legend(title = y_variable)
    plt.show()

y_variables = [col for col in cgf.columns if col not in ['Product', 'Income']]

y_variable_dropdown = widgets.Dropdown(
    options = y_variables,
    value = 'Age',
    description = ' y variable:'
)

output = widgets.interactive_output(update_chart, {'y_variable': y_variable_dropdown})
display(widgets.HBox([y_variable_dropdown]))
display(output)
```
