```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="darkgrid")
```

```python
diabetes=pd.read_csv("Diabetes.csv")
diabetes.head()
```

```python
sns.scatterplot(x=diabetes['Age'],y=diabetes['Insulin'])
plt.show()
```

```python
sns.relplot(data=diabetes, x='Age',y='Insulin',hue='Pregnancies',size='Pregnancies',alpha=.5)
plt.show()
```

```python
sns.relplot(data=diabetes, x='Age',y='Insulin',hue='Pregnancies',size='Pregnancies',sizes=(10,50))
plt.xlim(18,37)
plt.ylim(50.140)
plt.show()
```

This type of plot allows us to visualize the relationships between all pairs of continuous variables in the dataset, which can be helpful for identifying patterns and correlations.
